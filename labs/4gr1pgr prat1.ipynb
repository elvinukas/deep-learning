# Vaizdų klasifikavimas

```python
# Reikalingi paketai
import torch
import numpy as np
import torchvision
from torchvision import models
from PIL import Image
```

```python
# Patikrinama, ar yra GPU
print(torch.cuda.is_available())
device = torch.device('cuda:0')
print(device)
```

```python
# Modelio atsisiuntimas
model = models.vgg16(weights = models.VGG16_Weights.DEFAULT).to(device)
```

```python
# Transformacijos
transforms = models.VGG16_Weights.DEFAULT.transforms()
# transforms = torchvision.transforms.Compose([
#     torchvision.transforms.Resize((256, 256)),
#     torchvision.transforms.CenterCrop(224),
#     torchvision.transforms.ToTensor(),
#     torchvision.transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225)),
# ])
```

```python
# Vaizdo nuskaitymas iš disko
img = Image.open('dog.jpg')
```

```python
# Modelio paleidimas
logits = model(transforms(img).unsqueeze(0).to(device)).detach().cpu().numpy()
print(logits.shape)
logits = logits.squeeze()
label = np.argmax(logits)
print(label)
```

```python
# Žmogui suprantama klasė
# !wget https://klevas.mif.vu.lt/~mif28413/gmm/imagenet_classes.txt
with open('imagenet_classes.txt', 'r') as fid:
  all_classes = fid.readlines()
print(all_classes[label])
```
